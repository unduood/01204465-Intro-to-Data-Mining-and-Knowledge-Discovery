```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#Read Wine dataset, drop label column
df = pd.read_excel("Wine_66_2.xls", sheet_name="Data")
df.head()
```

```python
df.drop(['Type'], axis=1, inplace=True)
df.head()
```

```python
#Normalize dataset (X=unnormalized, X_norm=normalized)
X = df.values
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)
```

# <font face="Lato" size=5 color=orange>PCA with Unnormalized data</font>

```python
pca = PCA()
pca.fit(X)
```

## EigenVector

```python
#EigenVector
df_component = pd.DataFrame(pca.components_)
df_component.columns = df.columns
df_component = df_component.T
df_component.columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13']
df_component
```

## Variance ratio

```python
#Variance ratio
df_var = pd.DataFrame(pca.explained_variance_ratio_*100)
df_var.columns = ['Variance']
df_var = df_var.T
df_var.columns = df_var.columns+1
df_var
```

## a) Consider the row near the bottom labeled “variance”. Explain why column 1’s variance is so much greater that of any other column?

คอลัมน์ Proline มีค่าที่สูงกว่าคอลัมน์อื่นๆมากเมื่อ variance จึงมีค่าสูงขึ้น

## b) Comment on the use of normalization?

การทำ Normalization ทำให้ค่าในคอลัมน์ต่างๆอยู่บนสเกลใกล้กัน ช่วยป้องกันการ dominance จากบางคอลัมน์

## Plot cummulative variance

```python
#Plot cummulative variance
plt.figure(figsize=(10, 5))
plt.title('Cumulative Explained Variance as a Function of the Number of Components',fontsize=14)
plt.ylabel('Cumulative Explained variance')
plt.xlabel('Principal components')
cum_var = np.cumsum(df_var.values)
plt.plot(cum_var, color='r', marker='o')
plt.show()
```

กราฟ Cumulative Variance หลังจากทำการ normalize แสดงการเติบโตที่น้อยกว่ากราฟที่ยังไม่ได้ทำการ normalize อย่างชัดเจน เนื่องจากหลังการ normalize ค่า variance ของแต่ละองค์ประกอบจะลดลง ซึ่งทำให้การสะสมของค่า cumulative variance เกิดขึ้นช้ากว่าการวิเคราะห์ที่ยังไม่ผ่านการ normalize

## d) Without using PCA, create a classification model to predict Type of wine.


# Pre-process data

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df_with_type = pd.read_excel("Wine_66_2.xls", sheet_name="Data")

X = df_with_type.drop(['Type'], axis=1)
y = df_with_type['Type']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

## - Use 5-fold cross validation to evaluate performance of model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

# Perform 5-fold cross-validation
scores = cross_val_score(model, X_scaled, y, cv=5)

print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

## e) With using PCA, create a classification model to predict Type of wine.


## - Pre-process data


```python
X = df_with_type.drop(['Type'], axis=1)
y = df_with_type['Type']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

## - Explain setting of PCA parameters


n_components = 0.95 ใน PCA เป็นการกำหนดให้เลือกจำนวน principal components ที่สามารถอธิบายข้อมูลได้ถึง 95% ของ variance ทั้งหมด วิธีนี้ช่วยลดจำนวนมิติของข้อมูลลง แต่ยังคงรักษาข้อมูลสำคัญไว้อย่างครบถ้วน

## - Use 5-fold cross validation to evaluate performance of model

```python
# Create a classification model
model = LogisticRegression(max_iter=1000)

# Use 5-fold cross validation to evaluate performance of model

model.fit(X_train, y_train)
scores = cross_val_score(model, X_pca, y, cv=5)

print("Cross-validation scores:", scores)
print("Mean accuracy:", scores.mean())
```

## f) Discuss and Compare accuracy obtained between with PCA and without PCA


ความแม่นยำระหว่างการใช้ PCA และไม่ใช้ PCA มีค่าใกล้เคียงกันมาก (ประมาณ 0.988) ซึ่งบ่งชี้ว่า

ข้อมูลเดิมได้รับการเตรียมมาอย่างดีแล้ว ไม่มีความซ้ำซ้อนหรือสัญญาณรบกวนมากนัก ทำให้ PCA ไม่ได้ช่วยเพิ่มความแม่นยำอย่างมีนัยสำคัญ

PCA ช่วยลดจำนวนมิติของข้อมูล ส่งผลให้โมเดลทำงานได้เร็วขึ้น โดยไม่สูญเสียข้อมูลสำคัญ

หากข้อมูลมีมิติสูงและมีความเสี่ยงที่จะเกิดการ overfitting การใช้ PCA จะช่วยลดความซับซ้อนของโมเดล
