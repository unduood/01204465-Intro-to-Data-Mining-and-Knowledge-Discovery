```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
product_sale=pd.read_excel("PRODUCT_SALE.xlsx")
stock_name=pd.read_excel("STOCKNAME.xlsx")
```

```python
product_sale.head()
```

```python
stock_name.head()
```

```python
product_sale = product_sale.rename(columns={'BOUGHT_ID': 'CODE'})
product_sale = product_sale.merge(stock_name, on='CODE', how='left')
product_sale.drop('CODE', axis=1, inplace=True)
```

```python
product_sale.head()
```

```python
basket=product_sale.pivot_table(index='CUSTOMER_ID', columns='ITEMNAME', values='Quantity', aggfunc='sum').fillna(0)

basket=basket.astype(bool)
```

```python
basket
```

```python
frequent_itensets = apriori(basket, min_support=0.8, use_colnames=True)
frequent_itensets
```

```python
rules = association_rules(frequent_itensets, metric="lift", min_threshold=1)
rules.sort_values('support', ascending = False)
rules
```
