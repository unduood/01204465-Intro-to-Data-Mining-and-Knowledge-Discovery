## Importing the libraries

```python
#from sklearn.preprocessing import StandardScaler
#from sklearn.model_selection import train_test_split
#from sklearn.linear_model import LogisticRegression
#import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
```

#Prepare the data

```python
data = {
    'Color': ['Red', 'Red', 'Red', 'Yellow', 'Yellow', 'Yellow', 'Yellow', 'Yellow', 'Red', 'Red'],
    'Type': ['Sport', 'Sport', 'Sport', 'Sport', 'Sport', 'SUV', 'SUV', 'SUV', 'SUV', 'Sport'],
    'Origin': ['Domestic', 'Domestic', 'Domestic', 'Domestic', 'Imported', 'Imported', 'Imported', 'Domestic', 'Imported', 'Imported'],
    'Stolen': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes']
}

df = pd.DataFrame(data)
df.head(10)
```

#Set roles X and y for supervised-learning

```python
y_df = df[['Stolen']]
df.drop(['Stolen'], axis=1, inplace=True)
X_df = df
with pd.option_context('display.max_rows', 10): display(X_df)
```

```python
# .values attribute is used to convert the DataFrame y_df into a NumPy array.
X = X_df.values
y = y_df.values
print(X, '\n')
print(y)
```

#Encoding the Independent Variables (Predictor)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), list(range(X.shape[1])))], remainder='passthrough') # list(range(X.shape[1])) can change to [0,1,2]
X = np.array(ct.fit_transform(X))
print(X)
```

#Encoding the Dependent Variables (Target)

```python
from sklearn.preprocessing import LabelEncoder
y = y.ravel()  # or use y = y.reshape(-1,)
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

#Construct model

```python
k = 3
model = KNeighborsRegressor(n_neighbors=k)
model.fit(X, y)
```

#Predict unseen

```python
color_for_pred = "Red"
type_for_pred = "SUV"
origin_for_pred = "Domestic"
x_pred = [[1, 0, 1, 0, 1, 0]]
#x_pred_sc = scaler.transform(x_pred)
y_pred = model.predict(x_pred)
stolen_pred = "Yes" if y_pred[0] >= 0.5 else "No"

print('K =', k)
print(f'Color: {color_for_pred}, Type: {type_for_pred}, Origin: {origin_for_pred} => Stolen: {stolen_pred}')
print('Accuracy:', '{:.2f}'.format(model.score(X, y)))

''' Code from other colabs
k = 3
h = x_pred[0][0]
a = x_pred[0][1]
w = '{:.2f}'.format(y_pred[0])

print('K =', k)
print(f'Color: {h}, Age: {a} => Weight: {w}')
print('Accuracy:', '{:.2f}'.format(model.score(x, y)))
'''
```
