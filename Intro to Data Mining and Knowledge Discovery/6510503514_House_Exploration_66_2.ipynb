```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
## Display all the columns of the dataframe
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
## print shape of dataset with nb-rows and nb-columns

# Print the shape of the dataset
print(dataset.shape)
# Print the data types of each column
print(dataset.dtypes)
```

## Missing Values

```python
## Use heatmap tp show missing values features

sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.show()
```

```python
## Here we will check the percentage of nan values present in each feature
## 1 -step make the list of features which has missing values
## 2- step print the feature name and the percentage of missing values

features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

#### Temporal Variables(Eg: Datetime Variables)

```python
# list of variables that contain year information => Temporal variables
temporal_features = [feature for feature in dataset.columns if 'Yr' in feature or 'Year' in feature]
print(temporal_features)
```

```python
# let's explore the content of these year variables
for feature in temporal_features:
    print(f"{feature} {dataset[feature].unique()}")
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price
columns_of_interest = ['YrSold', 'SalePrice']
datasubset = dataset[columns_of_interest]
median_house_price = datasubset.groupby('YrSold')['SalePrice'].median().reset_index()
sns.lineplot(data=median_house_price, x='YrSold', y='SalePrice')
print(plt.title("House Price vs YearSold"))
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.show()
```

```python
## Numerical variables are usually of 2 type, Continous variable and Discrete Variables (finite set of values < 25)
import pandas as pd

# List of numerical variables
numerical_vars = dataset.select_dtypes(include=['int64', 'float64']).columns

# Distinguish discrete variables
discrete_feature = []

for var in numerical_vars:
    if var == 'YrSold':
        continue
    if dataset[var].nunique() < 25:
        discrete_feature.append(var)

#print("Discrete variables:", discrete_feature)
print("Discrete Variables Count:", len(discrete_feature))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

```python
## Lets Find the realtionship between Discrete Variables and Sale PRice
def plt_bar( col):
    median_col = dataset.groupby(col)['SalePrice'].median().reset_index()
    sns.barplot(data=median_col, x=col, y="SalePrice", width=0.5)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('SalePrice')
    plt.xticks(rotation=90)
    plt.show()

for col in discrete_feature:
    plt_bar( col)
```

#### Continuous Variable

```python
continuous_feature = []

for column in numerical_vars:
  if column not in temporal_features and column not in discrete_feature and column != 'Id':
    continuous_feature.append(column)

#print("Continuous variables:", continuous_feature)
print(f"Continuous Variables Count: {len(continuous_feature)}")
```

```python
## Lets analyse the continuous values by creating histograms to understand the distribution
def show_histograms(column):
    datasubset = dataset[column]
    sns.histplot(data=datasubset, bins=25, element='step')
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()

for column in continuous_feature:
    show_histograms(column)
```
