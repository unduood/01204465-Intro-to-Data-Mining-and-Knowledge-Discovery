# Import Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Read Dataset

```python
df = pd.read_csv('Cosmetics.csv')
df
```

# Pre-process

```python
df = df.astype(bool)
df
```

```python
df.drop('Trans. ', axis=1, inplace=True)
df
```

# Generate all the frequent itemsets with support = 20%

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets
```

# What is the total number of frequent itemsets?

```python
len(frequent_itemsets)
```

# Show all the maximal frequent itemsets.

```python
frequent_itemsets['Item Set Site'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.sort_values('support', ascending=False)
```

```python
frequent_itemsets.drop('Item Set Site', axis=1, inplace=True)
```

# Generate all the association rules with minimum support = 20% and minimum confidence = 70%

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.7)
rules
```

# Show rule with highest confidence

```python
rules.sort_values(by='confidence', ascending=False)
```

- Explain the meaning of this rule
 - เมื่อมีการซื้อ Mascara มีความเป็นไปได้ถึง 89.92% ที่ลูกค้าจะซื้อ Eye shadow เพิ่มด้วย ซึ่งบ่งชี้ว่าลูกค้ามักจะเลือกซื้อสินค้าทั้งสองรายการนี้พร้อมกัน
- Explain how support, confidence and lift are calculated?
 - Support: หมายถึงสัดส่วนของธุรกรรมทั้งหมดที่มีสินค้าทั้งสองรายการอยู่ร่วมกัน ยกตัวอย่างเช่น Support 32.1% หมายความว่า 32.1% ของธุรกรรมทั้งหมดประกอบด้วยทั้ง Mascara และ Eye shadow
 - Confidence: คือระดับความเชื่อมั่นของกฎ โดยคำนวณจากจำนวนธุรกรรมที่มีสินค้าทั้งสองรายการร่วมกัน หารด้วยจำนวนธุรกรรมที่มี Mascara
 - Lift: เป็นการประเมินว่าความน่าจะเป็นที่สินค้าสองรายการจะถูกซื้อพร้อมกันสูงเพียงใดเมื่อเทียบกับการซื้อแยกกันอย่างอิสระ
