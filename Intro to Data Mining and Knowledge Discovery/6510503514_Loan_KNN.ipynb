## Pre-process the dataset

```python
y_df = df[['Personal Loan']]
df.drop(['Personal Loan'], axis = 1, inplace = True)
X_df = df
with pd.option_context('display.max_columns', 8): display(X_df)
```

```python
X = X_df.values
y = y_df.values
```

```python
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)
new = new_customer.values
new = scaler.transform(new)
```

#K-NN k=3

```python
neigh = KNeighborsClassifier(n_neighbors=3, metric = 'euclidean')
neigh.fit(X_norm, y)
```

```python
y_pred = neigh.predict(new)
print(y_pred)
```

#Find best KNN

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=1992)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
scores = []
for k in range(1, 10):
    neigh = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
    neigh.fit(X_train, y_train.ravel())
    scores.append([k, neigh.score(X_test, y_test)])
```

```python
scores
```

```python
scores.sort(key=lambda x: x[1], reverse=True)
print(f"Best K= {scores[0][0]} with accuracy score: {scores[0][1]}")
```

#Using Best k & Show Confusion Matrix

```python
k = scores[0][0]
neigh = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
neigh.fit(X_train, y_train.ravel())
```

```python
y_pred = neigh.predict(X_test)
cn = confusion_matrix(y_test, y_pred)
print(cn)
print(classification_report(y_test, y_pred))
```
